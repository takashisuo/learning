```python
import pandas as pd
train = pd.read_csv('/content/drive/Othercomputers/DATACANVAS/github/learning/learning/samurai/ds_cource/23_kaggle/train.csv')
test = pd.read_csv('/content/drive/Othercomputers/DATACANVAS/github/learning/learning/samurai/ds_cource/23_kaggle/test.csv')
display(train.head())
display(test.head())
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
train=train.replace({'male': 0 ,'female': 1 })
test=test.replace({'male': 0 ,'female': 1 })

# "Age"がnullのレコードの"Age"カラムに対して、PCClass毎の中央値を適用する
# transformメソッドは、グループごとの計算結果を元のDataFrameの形状に合わせて返すため、データの前処理や特徴量エンジニアリングにおいて非常に有用
train.loc[train['Age'].isnull(), 'Age'] = train.groupby('Pclass')['Age'].transform('median')
test.loc[test['Age'].isnull(), 'Age'] = test.groupby('Pclass')['Age'].transform('median')
```

```python
# 基本の前処理
train_rev = train.copy()
train_rev['Embarked'] = train_rev['Embarked'].fillna(train['Embarked'].mode()[0])
test['Fare'] = test['Fare'].fillna(test['Fare'].median())
display(test['Fare'].median())
display(train_rev.head())
```

```python
# 参考（mode）
# サンプルデータの作成 data = pd.Series(['S', 'C', 'Q', 'S', 'C', 'S', 'Q', 'C', 'S', 'C']) # mode()メソッドの実行 result = data.mode()
import pandas as pd
data = pd.Series(['S', 'C', 'Q', 'S', 'C', 'S', 'Q', 'C', 'S', 'C'])
result = data.mode()
display(result)

data = pd.Series(['S', 'C', 'Q', 'S', 'C', 'S', 'Q', 'C', 'S', 'C', "S"])
result = data.mode()
display(result)

```

```python
# 名前から新規列の作成
pre_train = train_rev.copy()
pre_train['Title_tr'] = pre_train['Name'].map(lambda x: x.split(', ')[1].split('. ')[0])
pre_train['Title_tr'] = pre_train['Title_tr'].replace(['Dr','Rev','Mlle','Major','Col','the Countess','Capt','Ms','Sir','Lady','Mme','Don','Jonkheer'], 'Others')
pre_train['Title_tr'] = pre_train['Title_tr'].map( {'Master': 0 , 'Miss':1 , 'Mr':2, 'Mrs':3, 'Others':4})
```

```python
# testデータの作成
test['Title_te'] = test['Name'].map(lambda x: x.split(', ')[1].split('. ')[0])
test['Title_te'] = test['Title_te'].replace(['Col','Rev','Ms','Dr','Dona'], 'Others')
test['Title_te'] = test['Title_te'].map( {'Master': 0 , 'Miss':1 , 'Mr':2, 'Mrs':3, 'Others':4})
```

```python
# one-hot encoding のカラム結合
train_embarked_dum = pd.get_dummies(train['Embarked'], dtype='uint8')
display(train_embarked_dum.head())
train_add_dum = pd.concat([pre_train,train_embarked_dum], axis=1 )
display(train_add_dum.head())

test_embarked_dum = pd.get_dummies(test['Embarked'], dtype='uint8')
test_add_dum = pd.concat([test,test_embarked_dum], axis=1 )
display(test_add_dum.head())

```
