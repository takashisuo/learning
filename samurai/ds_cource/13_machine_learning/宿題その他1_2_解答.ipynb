````python

1. 2つのDataFrameの縦方向結合:

```python
import pandas as pd

df1 = pd.DataFrame({'A': [1, 2], 'B': [3, 4]})
df2 = pd.DataFrame({'A': [5, 6], 'B': [7, 8]})

result = pd.concat([df1, df2])
print(result)
```

2. 2つのDataFrameのマージ:

```python
df3 = pd.DataFrame({'id': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
df4 = pd.DataFrame({'id': [1, 2, 4], 'age': [25, 30, 35]})

# inner join
inner_merge = pd.merge(df3, df4, on='id', how='inner')
print("Inner Join:")
print(inner_merge)

# left join
left_merge = pd.merge(df3, df4, on='id', how='left')
print("\nLeft Join:")
print(left_merge)

# outer join
outer_merge = pd.merge(df3, df4, on='id', how='outer')
print("\nOuter Join:")
print(outer_merge)
```

3. 異なるキーでのマージ:

```python
df3 = pd.DataFrame({'id': [1, 2, 3], 'name': ['Alice', 'Bob', 'Charlie']})
df4 = pd.DataFrame({'no': [1, 2, 4], 'age': [25, 30, 35]})

left_merge = pd.merge(df3, df4, left_on='id', right_on='no', how='left')
print(left_merge)
```

4. カテゴリごとの平均計算:

```python
df5 = pd.DataFrame({
    'category': ['A', 'B', 'A', 'B', 'A', 'C'],
    'value': [10, 20, 30, 40, 50, 60]
})

result = df5.groupby('category')['value'].mean()
print(result)
```

5. ピボットテーブルの作成:

```python
df6 = pd.DataFrame({
    'date': ['2023-01-01', '2023-01-01', '2023-01-02', '2023-01-02'],
    'product': ['A', 'B', 'A', 'B'],
    'sales': [100, 150, 200, 250]
})

pivot_table = df6.pivot(index='date', columns='product', values='sales')
print(pivot_table)
```

6. null値の処理:

```python
import numpy as np

df7 = pd.DataFrame({
    'A': [1, 2, np.nan, 4, 5],
    'B': [5, np.nan, np.nan, 3, 2],
    'C': [1, 2, 3, 4, 5]
})

# 6-1: null値を含む行を全て削除
df7_1 = df7.dropna()
print("6-1: Null値を含む行を削除")
print(df7_1)

# 6-2: 列Bのnull値だけを削除
df7_2 = df7.dropna(subset=['B'])
print("\n6-2: 列Bのnull値だけを削除")
print(df7_2)

# 6-3: null値を0で置換
df7_3 = df7.fillna(0)
print("\n6-3: Null値を0で置換")
print(df7_3)
```

7. グラフ作成:

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# サンプルデータの作成
np.random.seed(0)
df = pd.DataFrame({
    'group': ['A', 'B', 'C', 'D'] * 25,
    'value1': np.random.normal(10, 2, 100),
    'value2': np.random.normal(50, 10, 100)
})

# 7-1: ヒストグラム
plt.figure(figsize=(10, 6))
plt.hist(df['value1'], bins=15)
plt.title('Histogram of value1')
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.show()

# 7-2: 棒グラフ
plt.figure(figsize=(10, 6))
df.groupby('group')['value1'].mean().plot(kind='bar')
plt.title('Average value1 by group')
plt.xlabel('Group')
plt.ylabel('Average value1')
plt.show()

# 7-3: 折れ線グラフ
plt.figure(figsize=(12, 6))
df.head(20)[['value1', 'value2']].plot(kind='line')
plt.title('Value1 and Value2 trends (first 20 rows)')
plt.xlabel('Index')
plt.ylabel('Value')
plt.legend(['Value1', 'Value2'])
plt.show()

# 7-4: 箱ひげ図
plt.figure(figsize=(10, 6))
sns.boxplot(x='group', y='value1', data=df)
plt.title('Distribution of value1 by group')
plt.show()

# 7-5: 散布図
plt.figure(figsize=(10, 6))
sns.scatterplot(x='value1', y='value2', hue='group', data=df)
plt.title('Scatter plot of value1 vs value2')
plt.show()
```

8. フィルタ処理:

```python
import pandas as pd
import numpy as np

# サンプルデータの作成
np.random.seed(0)
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Henry'],
    'age': [25, 30, 35, 40, 28, 32, 38, 45],
    'salary': [50000, 60000, 75000, 90000, 55000, 65000, 80000, 100000],
    'department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT', 'Finance', 'IT']
})

# 8-1: 年齢が30歳以上で、給与が70000以上の従業員を抽出
result_8_1 = df[(df['age'] >= 30) & (df['salary'] >= 70000)]
print("8-1: 年齢が30歳以上で、給与が70000以上の従業員")
print(result_8_1)

# 8-2: IT部門に所属し、年齢が35歳未満の従業員を抽出
result_8_2 = df.query("department == 'IT' and age < 35")
print("\n8-2: IT部門に所属し、年齢が35歳未満の従業員")
print(result_8_2)
```

9. Shift-JISファイルの読み込み:

```python
import pandas as pd

# ファイルの作成と書き込み
with open('sample.txt', 'w', encoding='shift_jis') as f:
    f.write("山田太郎")

# ファイルの読み込みと出力
with open('sample.txt', 'r', encoding='shift_jis') as f:
    content = f.read()
    print(content)
```

10. strftimeの使用:

```python
import pandas as pd

timestamp = pd.Timestamp('2023-05-15 14:30:00')
formatted_date = timestamp.strftime("%Y年%m月%d日 %H時%M分")
print(formatted_date)
```

11. strptimeの使用:

```python
from datetime import datetime

date_string = "2023-04-01 14:30"
date_object = datetime.strptime(date_string, "%Y-%m-%d %H:%M")

print("年:", date_object.year)
print("月:", date_object.month)
print("日:", date_object.day)
print("時:", date_object.hour)
print("分:", date_object.minute)
```

12. Pivot_table with aggfunc:

```python
import pandas as pd
import numpy as np

# サンプルデータの作成
np.random.seed(42)
data = {
    '日付': pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'),
    '商品': np.random.choice(['A', 'B', 'C'], size=365),
    '地域': np.random.choice(['東京', '大阪', '名古屋'], size=365),
    '売上': np.random.randint(1000, 10000, size=365)
}
df = pd.DataFrame(data)

# データの確認
print(df.head())

# pivot_tableを使用して、商品と地域ごとの売上の平均、最大値、最小値を計算
result = pd.pivot_table(df, values='売上', index='商品', columns='地域',
                        aggfunc=[np.mean, np.max, np.min])
print("\n商品と地域ごとの売上の統計:")
print(result)
```
````

````python
import pandas as pd
import numpy as np

# サンプルデータの作成
np.random.seed(0)
df = pd.DataFrame({
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Henry'],
    'age': [25, 30, 35, 40, 28, 32, 38, 45],
    'salary': [50000, 60000, 75000, 90000, 55000, 65000, 80000, 100000],
    'department': ['IT', 'HR', 'IT', 'Finance', 'HR', 'IT', 'Finance', 'IT']
})

print("サンプルデータ:")
print(df)
print("\n")

# 問題1: 条件式を使ったフィルタリング
print("問題1: 条件式を使ったフィルタリング")
print("年齢が30歳以上で、給与が70000以上の従業員を抽出してください。")
print("ヒント: df[条件式]の形式を使ってください。")

# 解答1
print("\n解答1:")
result1 = df[(df['age'] >= 30) & (df['salary'] >= 70000)]
print(result1)

print("\n" + "="*50 + "\n")

# 問題2: queryメソッドを使ったフィルタリング
print("問題2: queryメソッドを使ったフィルタリング")
print("IT部門に所属し、年齢が35歳未満の従業員を抽出してください。")
print("ヒント: df.query()メソッドを使ってください。")

# 解答2
print("\n解答2:")
result2 = df.query("department == 'IT' and age < 35")
print(result2)

print("\n" + "="*50 + "\n")

# 補足: 結果の確認
print("補足: 各問題の結果を確認しましょう。")
print("問題1の結果の行数:", len(result1))
print("問題2の結果の行数:", len(result2))
````
