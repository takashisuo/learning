````python
import matplotlib.pyplot as plt
import numpy as np

# データを作成
# 0から10までの幅で配列を等間隔に100個作成する
x = np.linspace(0, 10, 100)
y = 2 + 2 * np.sin(2 * x)

# プロット
# 重要: plt.subplotsではサブプロットを１つ作成する。
# 複数作成する場合は、plt.subplots(nrows, ncols)で行と列を指定する。
# fig, axes = plt.subplots(nrows, ncols)：nrows×ncols個のサブプロットを生成
# fig: Figure: 描画領域全体
# ax: Axes: 個々の座標軸
fig, ax = plt.subplots()

# 個別の座標軸側に描画する
ax.plot(x, y, linewidth=2.0)

# 個別の座標軸側でX軸・Y軸の設定をする
ax.set(xlim=(0, 10), xticks=np.arange(0, 10),
       ylim=(-1, 5), yticks=np.arange(-1, 6))


plt.show()
````

Figureの中にサブプロットとしてaxesがある。

````python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3)
y = x**2

fig = plt.figure()

# add_subplot(L, M, N)形式
# 1行2列の1番目
ax1 = fig.add_subplot(1,2,1)
ax1.plot(x,y)

# add_subplot(LMN)形式
# 1行2列の2番目
ax2 = fig.add_subplot(122)
x2 = np.arange(10)
ax2.bar(x2, 2**x2)

 # 全体を描画
plt.show()
````

````python
# title
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(-3,3)
y = x**2

fig = plt.figure()

# add_subplot(L, M, N)形式
# 1行2列の1番目
ax1 = fig.add_subplot(1,2,1)
ax1.plot(x,y)
ax1.set_xlabel("x1")
ax1.set_ylabel("y1")
ax1.set_title("sin")

# add_subplot(LMN)形式
# 1行2列の2番目
ax2 = fig.add_subplot(122)
x2 = np.arange(10)
ax2.bar(x2, 2**x2)
ax2.set_xlabel("x2")
ax2.set_ylabel("y2")
ax2.set_title("bar")

fig.suptitle("all")
 # 全体を描画
plt.show()
````

## 5.5 ヒストグラムを表示しよう

````python
import numpy as np
import matplotlib.pyplot as plt

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x = np.random.normal(15, 5, 2000)
# ヒストグラムをプロット
plt.hist(x)
````

````python
import numpy as np
import matplotlib.pyplot as plt

# 平均が15, 標準偏差5の標準正規分布から2000個の乱数の生成する
x = np.random .normal(15, 5, 2000)
# ヒストグラムをプロット

fig = plt.figure()
axes = fig.add_subplot(111)
axes.hist(x, bins=30)
axes.set_xlim(-10, 40)
axes.set_xlabel("x")
axes.set_ylabel("y")
axes.set_title("histgram")
plt.show()
````
