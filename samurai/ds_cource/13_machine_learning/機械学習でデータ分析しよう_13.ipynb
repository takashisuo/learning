```python
df = pd.read_csv("/content/drive/Othercomputers/DATACANVAS/github/learning/learning/samurai/ds_cource/13_machine_learning/california_housing_cleansing.csv")
df.head()
```

```python
df = df.drop(columns = ['Unnamed: 0'])
df.head()
```

```python
df.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
```

```python
X = df.to_numpy()
scaler.fit(X)
```

```python
X_scaled  = scaler.transform(X)
X_scaled
```

## 4. データセットのクラスタリング

```python
from sklearn.cluster import KMeans
```

```python
model = KMeans(n_clusters=4, random_state=0)
```

```python
model.fit(X_scaled)
```

```python
display(model.labels_)
"""
# 以下でもい格納可能
labels = model.predict(X_scaled)
labels
"""
```

```python
df['クラスター'] = model.labels_
df.head()
```

```python
sns.countplot(x='クラスター', data=df)
```

```python
df.groupby("クラスター").mean()
```

```python
sns.scatterplot(x='経度', y='緯度', hue='クラスター', data=df, alpha=0.5)
```

## 4. 特定のクラスターに的を絞った深掘り

```python
df_cluster3 = df.query('クラスター == 3')
sns.scatterplot(x='経度', y='緯度', data=df_cluster3)
```

```python
sns.histplot(x="地域人口", data=df, bins=50)
```
