## 1. データの取り込み

```python
import pandas as pd
import seaborn as sns
```

```python
from sklearn.datasets import fetch_california_housing
dataset = fetch_california_housing()
display(dataset)
```

```python
df = pd.DataFrame(dataset.data, columns=dataset.feature_names)
display(df.head())
```

```python
df['Price'] = dataset.target
df.head()
```

### 築年数

```python
# 築年数の調査
df.query('築年数 == 52').shape
```

```python
# 築年数の除外
df = df[df['築年数'] != 52]
display(df.shape)
```

## 住宅価格

```python
df.query('住宅価格 == 5.000010').shape
```

```python
df = df[df['住宅価格'] != 5.000010]
```

### 部屋数　寝室数　世帯人数

```python
sns.stripplot(x='部屋数', data=df)
```

```python
sns.stripplot(x='寝室数', data=df)
```

```python
sns.stripplot(x='世帯人数', data=df)
```

```python
df['世帯数'] = df['地域人口']/df['世帯人数']
df.head()
```

```python
df['全部屋数'] = df['部屋数']*df['世帯数']
df['全寝室数'] = df['寝室数']*df['世帯数']
df.head()
```

```python
df['部屋数/人'] = df['全部屋数']/df['地域人口']
df['寝室数/人'] = df['全寝室数']/df['地域人口']
df.head()
```

```python
sns.scatterplot(x='部屋数/人', y='寝室数/人', data=df)
```

```python
df = df.drop(columns = ['部屋数', '寝室数', '世帯人数', '世帯数', '全部屋数', '全寝室数'], errors='ignore')
display(df.head())
```
