## 7.3 条件に合うデータを抽出しよう

```python
# 初回のみライブラリをインポートする
import pandas as pd

# csvファイルをデータフレームとして読み込む
path = '/content/drive/MyDrive/study/data'
df = pd.read_csv(path + "/sample_pandas_6.csv")
# 先頭から5行目までを表示する
df.head()
```

```python
df.describe()
```

```python
# カラム名ではない文字列は、ダブルクォーテーション（”）で囲む必要がある
df.query('商品番号 == "Z4WOOIYV"')
```

```python
df.query('単価 == 600')
```

```python
# 複数条件に合う行
display(df.query('在庫 <= 5'))
display(df.query('商品番号 == "8T7D5DQA" and 在庫 <= 5'))
```

```python
# テキスト外(8を含む)
display(df.query('商品名.str.contains("8")').head(3))
display(df[df["商品名"].str.contains("8")].head(3))
# テキスト外(商品番号で指定の値を含む)
display(df.query('商品番号 in ["1QJFO8QY", "S6RE8W6X"]').head(3))
display(df[df["商品番号"].isin(["1QJFO8QY", "S6RE8W6X"])].head(3))

# テキスト外(変数)
thr_price = 600
display(df.query('単価 >= @thr_price').head(3))
display(df[df["単価"] >= thr_price].head(3))

# テキスト外(カラム名を変数で指定する)
col_product = "商品名"
product_name = "1QJFO8QY"
display(df.query(f'{col_product}.str.contains(@product_name)').head(3))
```

## 7.4 分析しやすいようにデータを変形しよう

```python
df[["単価"]]
```

```python
def tax(x, val:float=1.10):
    return x * val

display(df['単価'].apply(tax))
```

```python
df['単価'] * 1.10
```

```python
display(type(df['発注日'].loc[0]))
display(df['発注日'].loc[0])
```

```python
df['発注日'] = pd.to_datetime(df['発注日'])
display(df.info())
```

```python
display(df.agg({'発注日':['max', 'min']}))

# 以下でも可能（全列に適用）
display(df.agg(['max', 'min']))
```

```python
def tax(x, tax:float=1.10):
    return x * tax

df_ = df.copy()
tax_series = df_['単価'].apply(tax)
display(tax_series)
tax_series.name = "単価（税込み）"
# axis=1と同じ, 1:columnsが覚えにくい人はcolumns を利用
pd.concat([df_, tax_series], axis="columns")
```

```python
# 以下も同様。実務ではこちらをよく利用する。
df_ = df.copy()
df_["単価（税込み）"] = df_['単価'].apply(tax)
display(df_.head())
```

```python
# テキスト外: 個人的には以下が丁寧でよい
display(df['単価'].apply(lambda x: tax(x)))
# テキスト外: 以下でもできる
#display(df['単価'] * 1.10)


# テキスト外: 良く利用するのは以下のようなケース
df_ = df.copy()
df_["単価_税込み"] = df_["単価"].apply(tax, args=(1.20, ))
display(df_.head())

# 以下でも同様
#df_["単価_税込み"] = df_["単価"].apply(tax)
display(df_.head(3))

# テキスト外: 以下も利用する(axisの使い方に注意)
def calc(price, count):
    return price * count

df_ = df.copy()
df_["注文金額"] = df_[["単価", "注文数"]].apply(lambda x: calc(x[0], x[1]), axis=1)
display(df_.head(3))
```
