```
import pandas as pd
```

* このデータフレームをIDで集約してください。集約する際、日付はカウント、スコアは平均にしてください。集約は課題で利用したgroupbyを参考にしてください。ポイントは日付とスコアで違う集約方法ということです。



````python
import pandas as pd
import numpy as np
import random
from datetime import datetime, timedelta

# 日付範囲を設定
start_date = datetime(2022, 1, 1)
end_date = start_date + timedelta(days=99)

# 日付データを生成
date_range = pd.date_range(start=start_date, end=end_date)

# idとscoreのデータを生成
data = {'day': [],
        'id': [],
        'score': []}

# ランダムなデータを生成
for date in date_range:
    for _ in range(4):  # idごとに4レコード生成
        data['day'].append(date)
        data['id'].append(random.choice(['A', 'B', 'C', 'D']))
        data['score'].append(random.randint(1, 100))

# DataFrameに変換
df = pd.DataFrame(data)

aggregated_df = df.groupby('id').agg({'day': 'count', 'score': 'mean'}).reset_index()

# 列名の変更
aggregated_df.columns = ['ID', '日付数', '平均スコア']

# 結果を表示
print(aggregated_df)
````

````python
# サンプルデータの作成
import pandas as pd
import numpy as np

# 乱数シード値の設定
np.random.seed(123)

# データフレームの作成
data = {
    'name': ['Alice', 'Bob', 'Charlie', 'David', 'Eve', 'Frank', 'Grace', 'Hannah', 'Ivy', 'Jack'],
    'age': np.random.randint(20, 61, size=10),
    'city': ['Tokyo', 'Osaka', 'Nagoya', 'Fukuoka', 'Sapporo', 'Sendai', 'Hiroshima', 'Kobe', 'Kyoto', 'Yokohama'],
    'salary': np.random.randint(300000, 800001, size=10)
}

df = pd.DataFrame(data)

# 問題1の解答例
print(df.query('age >= 30'))

# 問題2の解答例
print(df.query('city == "Tokyo" or city == "Osaka"'))

# 問題3の解答例
print(df.query('salary >= 500000 and age < 40'))

# 問題4の解答例
print(df.query('name.str.startswith("A") or salary >= 700000'))

# 問題5の解答例
print(df.query('name.str.len() <= 4 and age >= 25 and age <= 35'))
````

# 実践8（応用）
* pd.read_csvでcsvファイルを読み込む際、カラムの型をdtypeで指定できます。実際に指定して読み込んでください。
* 例: df = pd.read_csv(csv_file, dtype=dtype_dict)
* 上記のdtype_dictを辞書型で指定する。一部調査してみてください。実務ではdtypeの指定は頻繁に利用します。
* 参考) pandasで指定する一般的な型

    int64: 整数型
    float64: 浮動小数点型
    bool: 真理値型
    object: 文字列型
    datetime64 または timedelta[ns]: 日時型


````python
import pandas as pd

# CSVファイルのパス
csv_file = 'data.csv'

# カラムの型を指定する辞書
dtype_dict = {
    'id': 'int64',
    'name': 'string',
    'age': 'int64',
    'height': 'float64',
    'is_student': 'bool'
}

# read_csvでカラムの型を指定してデータを読み込む
df = pd.read_csv(csv_file, dtype=dtype_dict)

print(df.dtypes)
````
