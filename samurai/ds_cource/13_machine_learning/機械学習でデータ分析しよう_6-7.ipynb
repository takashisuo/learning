# 7章 pandasを使いこなそう

```python
# 初回のみライブラリをインポートする
import pandas as pd

# csvファイルをデータフレームとして読み込む
df = pd.read_csv("sample_pandas_6.csv")
# 先頭から5行目までを表示する
df.head()
```

```python
# カラム名ではない文字列は、ダブルクォーテーション（”）で囲む必要がある
df.query('商品番号 == "Z4WOOIYV"')
```

```python
df.query('単価 == 600')
```

```python
# 複数条件に合う行
display(df.query('在庫 <= 5'))
display(df.query('商品番号 == "8T7D5DQA" and 在庫 <= 5'))
```

```python
def tax(x):
    return x * 1.10

display(df['単価'].apply(tax))

# 個人的には以下が丁寧でよい
display(df['単価'].apply(lambda x: tax(x)))

# 以下でもできる
display(df['単価'] * 1.10)
```

```python
display(type(df['発注日'].loc[0]))
display(df['発注日'].loc[0])
```

```python
df['発注日'] = pd.to_datetime(df['発注日'])
display(df.info())
```

```python
display(df.agg({'発注日':['max', 'min']}))

# 以下でも可能（全列に適用）
display(df.agg(['max', 'min']))
```
