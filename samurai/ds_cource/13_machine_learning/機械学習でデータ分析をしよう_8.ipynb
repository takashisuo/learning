# 8章 scikit-learnの概要を理解しよう

## 1. サンプルデータの取得

```python
from sklearn.datasets import load_wine

dataset = load_wine()
dataset
```

```python
display(dataset.data)
```

```python
display(dataset.feature_names)
```

```python
import pandas as pd
df = pd.DataFrame(data=dataset.data, columns=dataset.feature_names)
df.head()
```

```python
display(dataset.target)
display(type(dataset.target))
```

```python
# ndarrayもpandasに格納できる
df['category'] = dataset.target
display(df.head())
display(df.describe())
display(df.shape)
```

## 2. サンプルデータの分割

```python
X = dataset.data
y = dataset.target
display(f"type x: {type(X)}")
display(f"type y: {type(y)}")
```

```python
from sklearn.model_selection import train_test_split
train_test_split(X, y, test_size=0.3, random_state=5)
# 以下のようにrandom_stateを指定しない場合、実行毎に結果が異なります。
# train_test_split(X, y, test_size=0.3)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
print(f"X: {X.shape}, X_train: {X_train.shape}, X_test: {X_test.shape}, y: {y.shape}, y_train: {y_train.shape}, y_test:{y_test.shape}")
```

## 3. 予測モデルのインスタンス化

```python
from sklearn.tree import DecisionTreeClassifier
model = DecisionTreeClassifier(random_state=3)
# 以下のようにrandom_stateを指定しない場合、最終的に生成される予測モデルが実行毎に異なる可能性があります。これは、モデルの評価結果や予測結果が実行毎に変わってしまう可能性があること意味します。
# model = DecisionTreeClassifier()
```

## 4. 予測モデルの学習

```python
model.fit(X_train, y_train)
```

## 5. 予測モデルの評価

```python
y_pred = model.predict(X_test)
display(y_pred)
```

```python
display(y_test)
```

```python
# 正解率: accuracy
from sklearn.metrics import accuracy_score
accuracy_score(y_test, y_pred)
```

```python
# 以下でも算出できる
model.score(X_test, y_test)
```
