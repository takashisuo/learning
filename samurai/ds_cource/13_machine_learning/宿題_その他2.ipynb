## 3. 集計

1. 以下のデータを作成してください。

```
import pandas as pd
import numpy as np

# サンプルデータの作成
np.random.seed(42)
data = {
    '日付': pd.date_range(start='2023-01-01', end='2023-12-31', freq='D'),
    '商品': np.random.choice(['A', 'B', 'C'], size=365),
    '地域': np.random.choice(['東京', '大阪', '名古屋'], size=365),
    '売上': np.random.randint(1000, 10000, size=365)
}

df = pd.DataFrame(data)

# データの確認
print(df.head())

```
