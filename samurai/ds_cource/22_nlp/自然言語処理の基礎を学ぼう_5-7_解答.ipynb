## 問題7-1. テキストの形態素解析と品詞タグ付け
* 与えられた日本語テキストをspaCy+ja_ginzaで形態素解析し、各単語の品詞を表示してください。
* テキストは自分で作成してもよいですし、text = "私は東京で働いています。"を利用しても構いません。


````python
import spacy
import ginza

nlp = spacy.load("ja_ginza")

def analyze_text(text):
    doc = nlp(text)
    for token in doc:
        print(f"単語: {token.text}, 品詞: {token.pos_}")

# 使用例
text = "私は東京で働いています。"
analyze_text(text)
````

## 問題7-2. 固有表現(spaCy, ginza)
* テキストから人名、組織名、地名などの固有表現を抽出してください。
* テキストは自由ですが、以下でも構いません
* text = "山田太郎はソニーで働いており、来週東京に出張する予定です。"

````python
import spacy
import ginza

nlp = spacy.load("ja_ginza")

def extract_entities(text):
    doc = nlp(text)
    for ent in doc.ents:
        print(f"テキスト: {ent.text}, ラベル: {ent.label_}")

# 使用例
text = "山田太郎はソニーで働いており、来週東京に出張する予定です。"
extract_entities(text)
````

## 問題7-3: 依存関係解析 (応用: テキスト外, spaCy, ja_giza)
* 文の依存関係を解析し、各単語の係り受け関係を表示してください。
    * 例:
        ```
        単語: 美しい
        係り先: 花
        依存関係: acl
        単語: 花
        係り先: 咲い
        依存関係: nsubj
        ...

        ```


* テキストは任意ですが、以下でも構いません
* "美しい花が庭に咲いています。"
* ヒント: 係り受けを利用する問題です。以下のサイト等を参考にしてください。
* https://note.com/dsmemo/n/n159e8c96c8f4
    * ・解説①まずは書いてみる 付近のheadなど

````python
import spacy
import ginza

nlp = spacy.load("ja_ginza")

def analyze_dependencies(text):
    doc = nlp(text)
    for token in doc:
        print(f"単語: {token.text}")
        print(f"  係り先: {token.head.text}")
        print(f"  依存関係: {token.dep_}")

# 使用例
text = "美しい花が庭に咲いています。"
analyze_dependencies(text)
````
