```python
# 7.4 GPT-2で文章生成をしてみよう
# T5 (Text-To-Text Transfer Transformer): T5の大きな特徴は、すべてのNLPタスクを「入力テキストを受け取り、出力テキストを生成する」というフレームワークに統一している
from transformers import T5Tokenizer, AutoModelForCausalLM
tokenizer = T5Tokenizer.from_pretrained("rinna/japanese-gpt2-small")
model = AutoModelForCausalLM.from_pretrained("rinna/japanese-gpt2-small")

input = tokenizer.encode("自然言語処理とは", return_tensors="pt")
output = model.generate(input, do_sample=True, max_length=30, num_return_sequences=3)
print(tokenizer.batch_decode(output))
```

```python
# パラメータを変更した検証(max_length, temperature)
# temperature: サンプリングの温度。温度が低いと生成がより確定的になり、高いとよりランダムになります。通常は1.0ですが、0.7や1.2などで調整します。
input = tokenizer.encode("自然言語処理とは", return_tensors="pt")
output = model.generate(input, do_sample=True, max_length=100, num_return_sequences=1, temperature=1.2)
print(tokenizer.batch_decode(output))
```

```python
# 別のモデル: small -> medium
from transformers import T5Tokenizer, AutoModelForCausalLM
tokenizer = T5Tokenizer.from_pretrained("rinna/japanese-gpt2-medium")
model = AutoModelForCausalLM.from_pretrained("rinna/japanese-gpt2-medium")

input = tokenizer.encode("自然言語処理とは", return_tensors="pt")
output = model.generate(input, do_sample=True, max_length=128, num_return_sequences=1)
print(tokenizer.batch_decode(output))
```
