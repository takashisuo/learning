# 14章 変数のスコープを理解しよう

````python
def show_user_name():
    # ローカル変数を定義する
    user_name = "侍太郎"

    # ローカルスコープの範囲内でローカル変数を使う
    print(user_name)

show_user_name()
````

````python
def show_user_name():
    # ローカル変数を定義する
    user_name = "侍太郎"

    # ローカルスコープの範囲内でローカル変数を使う
    print(user_name)

show_user_name()

# ローカルスコープの範囲外でローカル変数を使う（エラーが発生する）
print(user_name)
````

````python
user_name = "侍花子"

def show_user_name():
    # ローカル変数を定義する
    user_name = "侍太郎"

    # ローカルスコープの範囲内でローカル変数を使う
    print(user_name)

show_user_name()

# グローバルスコープの範囲内でグローバル変数を使う
print(user_name)
````

````python
user_name = "侍花子"

def show_user_name():
    # ローカル変数を定義する
     user_name = "侍太郎"
     # user_name = "侍太郎"

     # ローカルスコープの範囲内でローカル変数を使う
     # グローバルスコープの範囲外でグローバル変数を使う（エラーが発生する）
    print(user_name)

show_user_name()

# グローバルスコープの範囲内でグローバル変数を使う
print(user_name)
````
