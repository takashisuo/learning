```python
# 1. リスト内包表記: 基本
# 問題: 1から10までの数字のリストを作成し、各数字を2倍にしたリストを作成してください。

# 解答:
numbers = [i * 2 for i in range(1, 11)]
print(numbers)
```

```python
# 2. リスト内包表記: if
# 問題: 1から20までの数字のリストから、偶数のみを抽出したリストを作成してください。

# 解答:
even_numbers = [i for i in range(1, 21) if i % 2 == 0]
print(even_numbers)
```

```python
# 3. リスト内包表記: if-else
# 問題: 1から10までの数字のリストを作成し、偶数は'even'、奇数は'odd'に置き換えたリストを作成してください。

# 解答:
number_types = ['even' if i % 2 == 0 else 'odd' for i in range(1, 11)]
print(number_types)
```

```python
# 4. ファイル入出力: read
# 問題: 'sample.txt'というファイルがあり、その中身が "Hello, World!" だとします。
# このファイルの内容を読み込んで表示するプログラムを書いてください。

# 解答:
with open('sample.txt', 'r') as file:
    content = file.read()
    print(content)
```

```python
# 5. ファイル入出力: readline
# 問題: 'numbers.txt'というファイルがあり、各行に1つずつ数字が書かれている全10行のファイルがあるとします。
# 最初の3行を読み込んで表示するプログラムを書いてください。

# 解答:
with open('numbers.txt', 'r') as file:
    for i in range(3):
        line = file.readline().strip()
        print(line)
```

```python
# 6. ファイル入出力: write
# 問題: 'output.txt'というファイルに "Python is fun!" という文字列を書き込むプログラムを作成してください。

# 解答:
with open('output.txt', 'w') as file:
    file.write("Python is fun!")
```

```python
# 7. ファイル入出力: writelines
# 問題: リスト ['apple\n', 'banana\n', 'cherry\n'] の内容を 'fruits.txt' ファイルに書き込むプログラムを作成してください。

# 解答:
fruits = ['apple\n', 'banana\n', 'cherry\n']
with open('fruits.txt', 'w') as file:
    file.writelines(fruits)
```

```python
# 8. Try-except: 基本
# 問題: ユーザーに数字の入力を求め、その数字を2で割った結果を表示するプログラムを作成してください。
# ただし、ユーザーが数字以外を入力した場合はエラーメッセージを表示してください。

# 解答:
try:
    number = int(input("数字を入力してください: "))
    result = number / 2
    print(f"結果: {result}")
except ValueError:
    print("エラー: 有効な数字を入力してください。")
```

```python
# 9. Try-catch: finally
# 問題: ファイル 'data.txt' を開いて内容を読み取り、表示するプログラムを作成してください。
# ファイルが存在しない場合はエラーメッセージを表示し、最後に必ず "処理が完了しました" と表示してください。

# 解答:
try:
    with open('data.txt', 'r') as file:
        content = file.read()
        print(content)
except FileNotFoundError:
    print("エラー: ファイルが見つかりません。")
finally:
    print("処理が完了しました")
```
