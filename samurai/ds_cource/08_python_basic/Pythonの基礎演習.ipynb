# 学生成績管理システム

あなたは学校の成績管理システムを作成することになりました。以下の要件を満たすPythonプログラムを作成してください。

1. 以下の学生と成績（3つのテスト結果）を持つ辞書 `grades` を作成してください：
   - 田中: [80, 75, 90]
   - 佐藤: [65, 70, 85]
   - 鈴木: [90, 95, 88]
   - 高橋: [72, 68, 81]

2. 成績の平均点を計算する関数 `calculate_average(scores)` を作成してください。

3. 合格した学生（平均点が特定の点数以上）をリストアップする関数 `get_passed_students(grades, passing_score)` を作成してください。

4. 新しい学生と成績を追加する関数 `add_new_student(grades, student, scores)` を作成してください。

5. 全生徒の成績と平均点を表示する関数 `print_all_grades(grades)` を作成してください。

6. メイン処理として、以下の操作を行ってください：
   a. 現在の全生徒の成績と平均点を表示する
   b. 合格点を75点とし、合格した学生をリストアップして表示する
   c. 新しい学生 "山田" と成績 [78, 82, 88] を追加する
   d. 更新後の全生徒の成績と平均点を表示する

注意：
- 関数を適切に使用してプログラムを構造化してください。
- コードにはコメントを付けて、各部分の目的を説明してください。
- 平均点は小数点第1位まで表示してください。

```python
def calculate_average(scores: list):
    """平均点を計算する関数"""
    return sum(scores) / len(scores)

def get_passed_students(grades: dict, passing_score: int) -> list:
    passed_students = []
    for student, scores in grades.items():
        if calculate_average(scores) >= passing_score:
            passed_students.append(student)
    return passed_students

def add_new_student(grades, student, scores):
    """新しい学生と成績を追加する関数"""
    grades[student] = scores
    print(f"{student}さんの成績を追加しました。")

def print_all_grades(grades):
    """全生徒の成績を表示する関数"""
    for student, scores in grades.items():
        average = calculate_average(scores)
        print(f"{student}: {scores}, 平均: {average:.1f}")

# メイン処理
grades = {
    "田中": [80, 75, 90],
    "佐藤": [65, 70, 85],
    "鈴木": [90, 95, 88],
    "高橋": [72, 68, 81]
}

print("現在の成績:")
print_all_grades(grades)

passing_score = 75
passed_students = get_passed_students(grades, passing_score)
print(f"\n合格した学生（{passing_score}点以上）:")
for student in passed_students:
    print(f"- {student}")

add_new_student(grades, "山田", [78, 82, 88])

print("\n更新後の全生徒の成績:")
print_all_grades(grades)
```
