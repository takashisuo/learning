```python
class Product:
    def __init__(self):
        self.name = ""

    # メソッドを定義する
    def set_name(self, name):
        self.name = name

    def show_name(self):
        print(self.name)

# インスタンス化する
coffee = Product()

# メソッドにアクセスして実行する
coffee.set_name("コーヒー")
coffee.show_name()
```

# 実践15-1:
* Personクラスを定義し、以下の条件を満たすようにしてください。
* コンストラクタでは、名前(name)と年齢(age)を受け取る
greet()メソッドを定義し、"Hello, my name is [名前]"と出力する
* get_year_of_birth()メソッドを定義し、現在年(2024年)から年齢を引いた値を返す

```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def greet(self):
        print(f"Hello, my name is {self.name}")

    def get_year_of_birth(self):
        current_year = 2024
        return current_year - self.age

# 使用例
person1 = Person("Alice", 25)
person1.greet()  # Hello, my name is Alice
print(f"Year of birth: {person1.get_year_of_birth()}")  # Year of birth: 1998
```

# 実践15-2:
* BankAccountクラスを定義し、以下の条件を満たすようにしてください。

* コンストラクタでは、口座番号(account_number)と残高(balance)を受け取る
* deposit(amount)メソッドを定義し、指定された金額を残高に加える
* withdraw(amount)メソッドを定義し、指定された金額を残高から引く(残高が不足している場合は"Insufficient balance"と出力する)
* get_balance()メソッドを定義し、現在の残高を返す

```python
class BankAccount:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposit of {amount} successful. New balance: {self.balance}")

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"Withdrawal of {amount} successful. New balance: {self.balance}")
        else:
            print("Insufficient balance")

    def get_balance(self):
        return self.balance

# 使用例
account1 = BankAccount("1234567890", 1000)
account1.deposit(500)  # Deposit of 500 successful. New balance: 1500
account1.withdraw(2000)  # Insufficient balance
account1.withdraw(1000)  # Withdrawal of 1000 successful. New balance: 500
print(f"Current balance: {account1.get_balance()}")  # Current balance: 500
```

# 実践15-3:
* Rectangleという以下のクラスを定義し、メソッドを呼び出してください。
* コンストラクタでは、幅(width)と高さ(height)を受け取る
calculate_area()メソッドを定義し、長方形の面積を計算して返す
* calculate_perimeter()メソッドを定義し、長方形の周囲長を計算して返す
* double_size()メソッドを定義し、幅と高さを2倍にする

```python
class Rectangle:
    def __init__(self, width, height):
        self.width = width
        self.height = height

    def calculate_area(self):
        return self.width * self.height

    def calculate_perimeter(self):
        return 2 * (self.width + self.height)

    def double_size(self):
        self.width *= 2
        self.height *= 2

# 使用例
rectangle1 = Rectangle(3, 4)
print(f"Area: {rectangle1.calculate_area()}")  # Area: 12
print(f"Perimeter: {rectangle1.calculate_perimeter()}")  # Perimeter: 14
rectangle1.double_size()
print(f"New area: {rectangle1.calculate_area()}")  # New area: 48
print(f"New perimeter: {rectangle1.calculate_perimeter()}")  # New perimeter: 28
```

# 実践15-4
* Studentクラスを定義し、以下の条件を満たすようにしてください。
* コンストラクタでは、名前(name)と年齢(age)を受け取る
add_grade(subject, grade)メソッドを定義し、受け取った科目(subject)と成績(grade)をdictionary形式で保存する
* get_grades()メソッドを定義し、保存された全ての科目と成績をdictionaryとして返す
* get_average_grade()メソッドを定義し、全ての科目の平均点を計算して返す

```python
class Student:
    def __init__(self, name, age):
        self.name = name
        self.age = age
        self.grades = {}

    def add_grade(self, subject, grade):
        self.grades[subject] = grade

    def get_grades(self):
        return self.grades

    def get_average_grade(self):
        total_grades = sum(self.grades.values())
        num_subjects = len(self.grades)
        if num_subjects > 0:
            average_grade = total_grades / num_subjects
            return average_grade
        else:
            return 0

# 使用例
student1 = Student("Alice", 20)
student1.add_grade("Math", 85)
student1.add_grade("Science", 92)
student1.add_grade("English", 78)

print(student1.get_grades())  # {'Math': 85, 'Science': 92, 'English': 78}
print(student1.get_average_grade())
```
