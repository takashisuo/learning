* 以下、最も利用するケース

```python
# datetimeモジュールをインポート
import datetime

# 現在時刻を取得する
now = datetime.datetime.now()

# 1年間の日数差を変数td_1yに代入する
td_1y = datetime.timedelta(days=365)

# 現在の日時に1年を加算し、変数addに代入する
add = now + td_1y

# 3日間の日数差を変数td_3dに代入する
td_3d = datetime.timedelta(days=3)

# 現在の日時から3日を減算し、変数subに代入する（sub＝「減算」を意味するsubtractionの略）
sub = now - td_3d

# 加算・減算した日時を特定のフォーマットで出力する
print(add.strftime("現在から1年後は%Y年%m月%d日%H時%M分%S秒です。"))
print(sub.strftime("現在から3日前は%Y年%m月%d日%H時%M分%S秒です。"))
```

# 以下も実務で利用する

```python
# datetimeモジュールをインポート
import datetime

# 現在時刻を取得する

now = datetime.datetime.now()

# 1年間の日数差を変数td_1yに代入する
td_1y = datetime.timedelta(days=-365)

# 現在の日時に1年を減算し、文字列として保持
diff = now + td_1y
expiration_date: str = diff.strftime("%Y%m%d")
print(expiration_date)

# 何かしら取得した%Y%m%dの値
target_date = "20230410"

if target_date < expiration_date:
    print("有効期限切れのためデータを削除します.")
```

# 総合問題2
* 銀行口座クラスBankAccountを作成してください
* initメソッドでは口座名義と残高を設定できるようにしてください
* depositメソッドでは引数として受け取った金額を残高に加算してください
* withdrawメソッドでは引数として受け取った金額を残高から引き、
* 残高が0未満にならないように設定してください
* また、withdrawメソッドが呼ばれるたびに「残高: xxxx円」と残高を出力するようにしてください。
* 出力のprintはfを用いてください。


```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"残高: {self.balance}円")
        else:
            print("残高が不足しています")

# 問題4
account = BankAccount("山田", 1000)
account.deposit(500)
account.deposit(500)
account.deposit(500)
account.withdraw(800)
account.withdraw(800)
```

# 総合問題4
* 標準ライブラリ time を用いて、時間がかかる処理を実装し、時間を計測してください。

```python
import time

def sum_numbers():
    total = 0
    for i in range(1, 1000001):
        total += i
    return total

# 処理開始前の時間を記録
start_time = time.time()

# 時間がかかる処理を実行
result = sum_numbers()
print(f"合計: {result}")

# 処理終了後の時間を記録
end_time = time.time()

# 実行時間を計算して出力
execution_time = end_time - start_time
print(f"実行時間: {execution_time:.6f}秒")
```

* この状況で、
* (1) dish_orderからメニュー番号を取り出し

* (3) 料理名を一つのリストに格納し、出力するコードを書いてください


dish_info = {1: "Curry", 2: "Pizza", 3: "Salad", 4: "Spaghetti"}

dish_order = [3, 1, 4, 1, 2]

```python
dish_info = {1: "Curry", 2: "Pizza", 3: "Salad", 4: "Spaghetti"}
dish_order = [3, 1, 4, 1, 2]

ordered_dishes = []
for order in dish_order:
    dish = dish_info[order]
    ordered_dishes.append(dish)

print(ordered_dishes)
```
