* withdrawメソッドでは引数として受け取った金額を残高から引き、

* 出力のprintはfを用いてください。


```python
class BankAccount:
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, amount):
        if self.balance >= amount:
            self.balance -= amount
            print(f"残高: {self.balance}円")
        else:
            print("残高が不足しています")

# 問題4
account = BankAccount("山田", 1000)
account.deposit(500)
account.deposit(500)
account.deposit(500)
account.withdraw(800)
account.withdraw(800)
```

# 総合問題17-3
* 標準ライブラリ time を用いて、時間がかかる処理を実装し、時間を計測してください。

```python
import time

def sum_numbers():
    total = 0
    for i in range(1, 1000001):
        total += i
    return total

# 処理開始前の時間を記録
start_time = time.time()

# 時間がかかる処理を実行
result = sum_numbers()
print(f"合計: {result}")

# 処理終了後の時間を記録
end_time = time.time()

# 実行時間を計算して出力
execution_time = end_time - start_time
print(f"実行時間: {execution_time:.6f}秒")
```

# 総合問題17-5: 従業員名簿の作成
会社の従業員名簿を作成するプログラムを作成してください。以下の仕様を満たすようにしてください。

Employeeクラスを作成する

属性: name(名前)、age(年齢)、department(所属部署)、salary(給与)
__init__メソッド(コンストラクタ)で上記属性を初期化する
raise_salaryメソッドを作成し、self.salaryに5%アップした値を設定する


従業員のリストemployeesを作成する
create_employee関数を作成する

引数としてname、age、department、salaryを受け取る
Employeeクラスのインスタンスを作成し、返す


create_employee関数を使って、employeesリストに5人の従業員を追加する

適当な名前、年齢、部署、給与を設定する


全従業員の名簿を出力する

従業員一人につき、name、age、department、salaryを出力する


35歳以上の従業員に5%の給与アップを実施する

employeesリストから35歳以上の従業員を探し出す
その従業員インスタンスのraise_salaryメソッドを呼び出す


新しい従業員名簿を出力する(給与アップ後の名簿)



```python
class Employee:
    def __init__(self, name, age, department, salary):
        self.name = name
        self.age = age
        self.department = department
        self.salary = salary

    def raise_salary(self):
        self.salary *= 1.05  # 5% 増加

def create_employee(name, age, department, salary):
    return Employee(name, age, department, salary)

# 従業員リストの作成
employees = []

# 従業員の追加
employees.append(create_employee("John Doe", 32, "Sales", 50000))
employees.append(create_employee("Jane Smith", 28, "Marketing", 60000))
employees.append(create_employee("Bob Johnson", 45, "Engineering", 80000))
employees.append(create_employee("Alice Williams", 38, "HR", 70000))
employees.append(create_employee("Tom Brown", 51, "Finance", 90000))

# 従業員名簿の出力
print("従業員名簿:")
for employee in employees:
    print(f"名前: {employee.name}, 年齢: {employee.age}, 部署: {employee.department}, 給与: {employee.salary}")

# 35歳以上の従業員に給与アップ
print("\n35歳以上の従業員に5%の給与アップを実施します。")
for employee in employees:
    if employee.age >= 35:
        employee.raise_salary()

# 新しい従業員名簿の出力
print("\n新しい従業員名簿:")
for employee in employees:
    print(f"名前: {employee.name}, 年齢: {employee.age}, 部署: {employee.department}, 給与: {employee.salary}")
```

この問題では、ユーザプロフィールを表すクラスを作成し、さらにそのクラスを加工するための関数を実装します。

UserProfileクラスを作成してください。このクラスには以下のプロパティを持たせます。

name (名前)
age (年齢)
email (メールアドレス)
bio (自己紹介文)


update_profile関数を作成してください。この関数はUserProfileインスタンスを引数として受け取り、以下の処理を行います。

nameプロパティを大文字に変換する
ageプロパティに5を加算する
emailプロパティに"@example.com"を追加する (既にドメインが含まれている場合は変更しない)
bioプロパティが空文字列の場合は"No bio provided."に設定する


update_profile関数から返されたUserProfileインスタンスを使って、以下のように出力してください。
Copy code名前: [名前]
年齢: [年齢]
メールアドレス: [メールアドレス]
自己紹介: [自己紹介文]

次のユーザプロフィールを使ってプログラムを実行し、結果を確認してください。
pythonCopy codeuser1 = UserProfile("John Doe", 30, "john@example.com", "I am a software engineer.")
user2 = UserProfile("jane_smith", 25, "janesmith", "")

```python
class UserProfile:
    def __init__(self, name, age, email, bio):
        self.name = name
        self.age = age
        self.email = email
        self.bio = bio

def update_profile(user):
    user.name = user.name.upper()
    user.age += 5
    if "@" not in user.email:
        user.email += "@example.com"
    if not user.bio:
        user.bio = "No bio provided."
    return user

# ユーザプロフィールの作成
user1 = UserProfile("John Doe", 30, "john@example.com", "I am a software engineer.")
user2 = UserProfile("jane_smith", 25, "janesmith", "")

# update_profile関数を呼び出し、加工されたインスタンスを取得
updated_user1 = update_profile(user1)
updated_user2 = update_profile(user2)

# 結果を出力
print("ユーザ1:")
print(f"名前: {updated_user1.name}")
print(f"年齢: {updated_user1.age}")
print(f"メールアドレス: {updated_user1.email}")
print(f"自己紹介: {updated_user1.bio}")

print("\nユーザ2:")
print(f"名前: {updated_user2.name}")
print(f"年齢: {updated_user2.age}")
print(f"メールアドレス: {updated_user2.email}")
print(f"自己紹介: {updated_user2.bio}")
```

```python
import re

class UserProfile:
    def __init__(self, name, age, email, bio):
        self.name = name
        self.age = age
        self.email = email
        self.bio = bio

def update_profile(user):
    user.name = user.name.upper()
    user.age += 5
    if not re.match(r".*@.+\..+", user.email):
        user.email += "@example.com"
    if not user.bio:
        user.bio = "No bio provided."
    return user

# ユーザプロフィールの作成
user1 = UserProfile("John Doe", 30, "john@example.com", "I am a software engineer.")
user2 = UserProfile("jane_smith", 25, "janesmith", "")

# update_profile関数を呼び出し、加工されたインスタンスを取得
updated_user1 = update_profile(user1)
updated_user2 = update_profile(user2)

# 結果を出力
print("ユーザ1:")
print(f"名前: {updated_user1.name}")
print(f"年齢: {updated_user1.age}")
print(f"メールアドレス: {updated_user1.email}")
print(f"自己紹介: {updated_user1.bio}")

print("\nユーザ2:")
print(f"名前: {updated_user2.name}")
print(f"年齢: {updated_user2.age}")
print(f"メールアドレス: {updated_user2.email}")
print(f"自己紹介: {updated_user2.bio}")
```

# 総合問題17-7: 日付計算
* 現在日から以下の条件を満たす日付を計算し、それぞれの日付を計算してください。
    * 2週間後の日付
    * 一カ月前の日付
    * 365日後の日付
* 上記のいずれかを文字列として変数に格納してください。

```python
from datetime import datetime, timedelta

# 現在日時を取得
current_date = datetime.now()

# 2週間後の日付を計算
two_weeks_later = current_date + timedelta(days=14)

# 1ヶ月前の日付を計算
one_month_ago = current_date - timedelta(days=30)

# 365日後の日付を計算
one_year_later = current_date + timedelta(days=365)

# 日付を文字列として変数に格納
date_str = str(one_month_ago.date())

print(f"現在日時: {current_date}")
print(f"2週間後の日付: {two_weeks_later}")
print(f"1ヶ月前の日付: {one_month_ago}")
print(f"365日後の日付: {one_year_later}")
print(f"文字列への変換例: {date_str}")
```

# 総合問題17-8(関数とクラスの応用): 関数渡し

関数の引数やクラスのメンバー(属性)に関数を渡すことができます。
クラスのインスタンス化時に関数を渡すことで、インスタンスごとに異なる動作を実現できます。
以下、問題です。

動物の鳴き声を出力するプログラムを作ってください。
しかし、動物の種類によって鳴き声は異なるので、動物ごとに鳴き声を出力する関数を用意するのは面倒です。そこで、動物クラスのインスタンスを生成する際に、鳴き声を出力する関数を渡せるようにしたいと思います。

* Animalクラスを作成してください
* 初期化時に鳴き声関数を引数で受け取り属性として設定してください
* make_soundメソッド内で鳴き声を鳴らしてください

```python
class Animal:
    def __init__(self, sound_func):
        """
        sound_func: 鳴き声を出力する関数
        """
        self.sound = sound_func

    def make_sound(self):
        """
        鳴き声を出力する
        """
        print(self.sound())

# 鳴き声関数の例
def dog_bark():
    return "ワンワン"

def cat_meow():
    return "ニャー"

# 使用例
dog = Animal(dog_bark)
cat = Animal(cat_meow)

dog.make_sound() # 出力: ワンワン
cat.make_sound() # 出力: ニャー
```
