````python
for i in range(1, 11):
    print(i)

    #ループ変数の値が5であれば、break文で繰り返し処理を強制終了する
    if i == 5:
        print(f"{i}: breakします")
        break
````

````python
import random

threshold = 20

for i in range(1, 11):
    #変数numに1~20までのランダムな整数を代入する
    num = random.randint(1,20)

    print(f"{i}回目の結果は{num}です。")

    #変数numの値が20であれば、break文で繰り返し処理を強制終了する
    if num == threshold:
        print(f"{threshold}が出たので繰り返し処理を強制終了します。")
        break
````

````python
for i in range(1, 11):
    # ループ変数iの値が奇数（2で割った余りが1）であれば、値を出力せずにcontinue文で次のループに進む
    if i % 2 == 1:
        continue

    print(i)
````

#10章 繰り返し処理のwhile分を理解しよう

````python
#ランダムな整数を利用するため、記述が必要です。
import random

# 変数numに0~4までのランダムな整数を代入する
num = random.randint(0, 4)

# 変数numの値が0以外である間、変数numの値を出力し続ける
while num != 0:
    num = random.randint(0, 4)
    print(num)
````

````python
#ランダムな整数を利用するため、記述が必要です。
import random

# 変数numに0~4までのランダムな整数を代入する
num = random.randint(0, 4)

# 変数numの最初の値を出力する（確認用）
print(f"最初の値は{num}です。")

# 変数numの値が0以外である間、変数numの値を出力し続ける
while num != 0:
    # 変数numに0~4までのランダムな整数を代入する
    num = random.randint(0, 4)

    # 次の条件式で比較される、変数numの現在の値を出力する
    print(f"現在の値は{num}です。")
````

````python
import random

# 変数numに0~4までのランダムな整数を代入する
num = random.randint(0, 4)

# カウンタ変数を用意する
i = 1

# 変数numの最初の値を出力する（確認用）
print(f"最初の値は{num}です。")

# 変数numの値が0以外である間、変数numの値を出力し続ける
while num != 0:
    # 変数numに0~4までのランダムな整数を代入する
    num = random.randint(0, 4)

    # カウンタ変数iの値が5であれば、break文で繰り返し処理を強制終了する
    if i == 5:
        print("5回目なので繰り返し処理を強制終了します。")
        break

    # 次の条件式で比較される、変数numの現在の値を出力する
    print(f"現在の値は{num}です。")
    #　カウンタ変数のの値を1増やす
    i = i + 1
````

````python
#ランダムな整数を利用するため、記述が必要です。
import random

# 合計用の変数を用意する
sum = 0

# 変数sumの値が20以上になるまで繰り返し処理を行う
while sum < 20:
    # 変数numに1~10までのランダムな整数を代入する
    num = random.randint(1, 10)

    print(f"{num}が出ました。")

    # 変数numの値が偶数（2で割った余りが0）であれば、加算せずにcontinue文で次のループに進む
    if num % 2 == 0:
        print("偶数なので加算しません。")
        continue

    # 変数sumに変数numの値を加算する
    sum = sum + num
    print(f"現在の合計は{sum}です。")
````

# 11章 配列や辞書の繰り返し処理を理解しよう

````python
user_names: list = ["侍太郎", "侍一郎", "侍二郎", "侍三郎", "侍四郎"]

for user_name in user_names:
    print(f"{user_name} です")
````

````python
personal_data: dict = {"name": "侍太郎", "age": 36, "gender": "男性"}

for key, value in personal_data.items():
    print(f"{key}は{value}です。")
````

````python
personal_data = {"name": "侍太郎", "age": 36, "gender": "男性"}

# valueのみ
for value in personal_data.values():
    print(value)

# keyのみ
for key in personal_data.keys():
    print(key)
````

````python
user_names = ["侍太郎", "侍一郎", "侍二郎", "侍三郎", "侍四郎"]

# 検索したいユーザー名を代入する変数
target = "侍二郎"

for user_name in user_names:
    print(user_name)

    # 変数user_nameと変数targetの値が一致すれば、break文で繰り返し処理を強制終了する
    if (user_name == target):
        print(f"{target}さんが見つかったので、繰り返し処理を強制終了します。")
        break
````

````python
score = {
    "国語": 80,
    "数学": 55,
    "理科": 70,
    "社会": 85,
    "英語": 60
}

print("合格した科目は以下のとおりです。")

for key, value in score.items():
    # 変数valueの値（点数）が70より小さければ、キー（科目名）と値（点数）を出力せずにcontinue文で次のループに進む
    if (value < 70):
        continue
    print(f"{key} : {value}点")
````

# 実践3
* 1から100までの数字を出力するプログラムを作成してください。ただし、3の倍数の場合は'Fizz'、5の倍数の場合は'Buzz'、3と5の公倍数の場合は'FizzBuzz'と出力してください。

````python
for i in range(1, 101):
    if i % 3 == 0 and i % 5 == 0:
        print("FizzBuzz")
    elif i % 3 == 0:
        print("Fizz")
    elif i % 5 == 0:
        print("Buzz")
    else:
        print(i)
````

# 実践4
* 文字列(英語)を入力として受け取り、その文字列内の母音の数を数えるプログラムを作成してください。ただし、大文字と小文字は区別しないものとします。

````python
input_string = input("Enter a string: ")
vowels = "aiueoAIUEO"
count = 0

for char in input_string:
    if char in vowels:
        count += 1
    else:
        continue

print(f"The number of vowels in '{input_string}' is {count}")
````

# 実践6
* 2つのリストをforで回し、それらをtupleとして新しいリストへ格納するプログラムを作成してください。

````python
list1 = [1, 2, 3, 4, 5]
list2 = ["c#", "vb", "java"]
new_list = []
for l1 in list1:
    for l2 in list2:
        new_list.append((l1, l2))

print(new_list)
````

# 実践7
* リストに数字を複数入力し、入力した数字の合計値が100を超えたら繰り返しを終了するプログラムを作成してください。inputとint変換を使うとわかりやすいです。

````python
# 空のリストを作成
numbers = []

# 合計値を初期化
total = 0

# 無限ループ
while True:
    # 数値入力を求める
    try:
        num = int(input("数値を入力してください: "))
    except ValueError:
        print("無効な入力です。数値を入力してください。")
        continue

    # リストに追加し、合計値を更新
    numbers.append(num)
    total += num

    # 合計値が100を超えた場合は終了
    if total > 100:
        break

# 入力した数値とその合計値を出力
print("入力した数値:", numbers)
print(f"入力回数: {len(numbers)}")
print("合計値:", total)
````

````python
import random
numbers = random.randint(1,50)
numbers_sum_list = []
numbers_sum_list.append(numbers)
numbers_sum = numbers

while numbers_sum <= 100:
    numbers = random.randint(1,50)
    numbers_sum = numbers_sum + numbers
    numbers_sum_list.append(numbers)

print(numbers_sum_list)
print(f"合計{numbers_sum}")
````

# 実践8
* 与えられたディクショナリーから、値が偶数のキーと値のペアのみを新しいディクショナリーとして出力するプログラムを作成してください。

````python
user_dict = {'apple': 2, 'banana': 3, 'cherry': 4, 'date': 6, 'elderberry': 8}
new_dict = {}

for key, value in user_dict.items():
    if value % 2 == 0:
        new_dict[key] = value
    else:
        continue

print("Original dictionary:", user_dict)
print("New dictionary with even values:", new_dict)
````
