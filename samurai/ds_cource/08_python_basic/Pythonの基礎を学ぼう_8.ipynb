```python
num = 15

# すべての条件が成り立つ場合にのみ処理を行う（変数numの値が11～29）
if 10 < num and num < 30:
    print("変数numは10より大きく、30より小さいです")

# 1つでも条件が成り立てば処理を行う（変数numの値が10または30）
if num == 10 or num == 30:
    print("変数numは10または30です")
```

# 課題: 条件分岐のif文で処理を切り替えよう

if文を用いて、以下の条件分岐をするコードを記述してください。値はprint関数で出力してください。

* 変数varが、3の倍数の場合は「Fizz」を出力
* 変数varが、5の倍数の場合は「Buzz」を出力
* 変数varが、3の倍数と5の倍数の両方に該当する場合は「FizzBuzz」を出力
* 上記のどの場合にも該当しない場合は、変数varの値を出力

# 課題後の実践1:
1. pythonの比較演算子('==', '!=', '>', '<', '>=', '<=')をすべて検証してください。
1. 例: 変数に値を入れる。その値と別の値を比較し、それぞれの比較演算子を満たす場合はprintを出力する。

```python
# 変数numに10を代入
num = 10

# 変数numが5より大きいかどうかを比較
if num > 5:
    print("変数numは5より大きいです")

# 変数numが15より小さいかどうかを比較
if num < 15:
    print("変数numは15より小さいです")

# 変数numが10と等しいかどうかを比較
if num == 10:
    print("変数numは10と等しいです")

# 変数numが20と等しくないかどうかを比較
if num != 20:
    print("変数numは20と等しくありません")

# 変数numが12以上かどうかを比較
if num >= 12:
    print("変数numは12以上です")

# 変数numが8以下かどうかを比較
if num <= 8:
    print("変数numは8以下です")
```

# 課題後の実践4:
1. 2つの数値を入力し、大きい方の数値を出力するプログラムを作成してください。ただし、2つの数値が同じ場合は"同じ値です"と出力してください。

```python
num1 = float(input("1つ目の数値を入力してください: "))
num2 = float(input("2つ目の数値を入力してください: "))

if num1 > num2:
    print(f"大きい方の数値は{num1}です")
elif num1 < num2:
    print(f"大きい方の数値は{num2}です")
else:
    print("同じ値です")
```

# 課題後の実践5:
- 西暦の年を入力し、その年がうるう年かどうかを判定するプログラムを作成してください。うるう年の条件は以下です。
    - 西暦年が4で割り切れる年はうるう年
    - ただし、100で割り切れる年はうるう年ではない
    - しかし、400で割り切れる年はうるう年 となります。
- ヒント: 条件文の優先事項が何かに注意してください。


```python
year = int(input("西暦年を入力してください: "))

if year % 400 == 0:
    print(f"{year}年はうるう年です")
elif year % 100 == 0:
    print(f"{year}年はうるう年ではありません")
elif year % 4 == 0:
    print(f"{year}年はうるう年です")
else:
    print(f"{year}年はうるう年ではありません")
```

```python
years = int(input("西暦年を入力してください: "))

if years % 4 != 0:
    print(f"西暦{years}年です。うるう年ではありません")

elif years % 100 == 0 and years % 400 != 0:
    print(f"西暦{years}年です。うるう年ではありません")
else:
    print(f"西暦{years}年です。うるう年です")
```
